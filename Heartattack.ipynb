### Resting electrocardiographic results

```python
df.restecg.value_counts()
```

```python
plt.figure(figsize=(5,5))
plt.pie(df.restecg.value_counts(),labels=df.restecg.value_counts().index.tolist(),autopct='%1.1f%%')
```

Resting electrocardiographic results:

### Chest Pain v/s Sex

### Chest Pain v/s Output

```python
chestpain_output = df.groupby(['cp', 'output']).size().unstack().fillna(0)
chestpain_output.plot(kind='bar', stacked=True)
plt.title('Chest Pain Type vs Output')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count of Patients')
plt.xticks(ticks=[0,1,2,3], labels=['typical angina', 'atypical angina', 'non-anginal pain', 'asymptomatic'], rotation=0)
plt.show()
```


*  Typical Anginal chest pain shows less chances of heart attack in patients.
*  Non-anginal chest pain shows high chances heart attack in patients.



### Chest pain v/s ECG

```python
cp_ecg = df.groupby(['cp', 'restecg']).size().unstack().fillna(0)
cp_ecg.plot(kind='bar', stacked=True)
plt.title('Chest Pain Type vs resting electrocardiographic results')
plt.xlabel('Chest Pain Type')
plt.ylabel('resting electrocardiographic results')
plt.xticks(ticks=[0,1,2,3], labels=['typical angina', 'atypical angina', 'non-anginal pain', 'asymptomatic'], rotation=0)
plt.show()
```

Chest pain likely to occur when BP count is more than 120 with chances of having ST-T wave abnormality & showing probable or definite left ventricular hypertrophy by Estes' criteria.

### Blood pressure v/s ECG 

```python
sns.histplot(data=df,x='trtbps',hue='restecg')
```

From the above visualtion we could notice that when BP counts more than 120, ECG shows probable or definite left ventricular hypertrophy by Estes' criteria, which is a condition where LVH is an independent risk factor for cardiovascular disease. Once LVH is developed, it puts the patient at significant risk of developing myocardial ischemia and infarction, heart failure, dysrhythmias, or even sudden death.

### Maximum heart rate & chest pain

```python
sns.scatterplot(data=df,x="cp",y='thalachh')
```

As and when there is high heart rate ,chest pain is likely to be more & the types of chest pain occur are atypical angina,non-anginal pain,asymptomatic.

where ,
  Typical angina =substernal chest pain precipitated by physical exertion or emotional stress and relieved with rest or nitroglycerin.
  Atypical angina= Angina pectoris which does not have associated classical symptoms of chest pain. Symptoms may include weakness, nausea, or sweating.
  Nonanginal chest pain = which doctors may also call noncardiac chest pain, refers to pain a person without heart disease may feel behind the breast bone.
  Asymptomatic pain= A transient alteration in myocardial perfusion in the absence of chest pain or the usual anginal equivalents.
