```python
Flightdist=df[['Gender','Customer Type','Flight Distance','Satisfaction']]
Flightdist.head(10)
```

```python
NewFlightdist = Flightdist[Flightdist['Customer Type']=='Returning'] # Copy the Flightdist data frame to a new dataframe where the customer type is equal to returning

NewFlightdist.head()
```

```python
# Change Flight Distance values where flight distance is less than 750 to 0

NewFlightdist["Flight Distance"] = np.where(NewFlightdist["Flight Distance"]<=750, 0,NewFlightdist["Flight Distance"])


# Change Flight Distance values to 1 when flight distance is greater than 750 but less than 1500.

NewFlightdist["Flight Distance"] = np.where((NewFlightdist["Flight Distance"]> 750) & (NewFlightdist["Flight Distance"] <=1500), 1,NewFlightdist["Flight Distance"])

# Change Flight Distance values to 2 when flight distance is greater than 1500 but less than 2250.

NewFlightdist["Flight Distance"] = np.where((NewFlightdist["Flight Distance"]> 1500) & (NewFlightdist["Flight Distance"] <=2250), 2,NewFlightdist["Flight Distance"])


# Pallu's To Do's
# Change Flight Distance values to 3 when flight distance is greater than 2250 but less than 3000.

NewFlightdist["Flight Distance"] = np.where((NewFlightdist["Flight Distance"]> 2250) & (NewFlightdist["Flight Distance"] <=3000),3,NewFlightdist["Flight Distance"])

# Change Flight Distance values to 4 when flight distance is greater than 3000 but less than 3750.
NewFlightdist["Flight Distance"] = np.where((NewFlightdist["Flight Distance"]> 3000) & (NewFlightdist["Flight Distance"] <=3750), 4,NewFlightdist["Flight Distance"])


# Change Flight Distance values to 5 when flight distance is greater than 3750.

NewFlightdist["Flight Distance"] = np.where((NewFlightdist["Flight Distance"]> 3750) , 5,NewFlightdist["Flight Distance"])
```
